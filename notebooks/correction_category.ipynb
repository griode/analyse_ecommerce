# Préparation et entraitement

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
def clean_text(text):
    """Nettoie et préprocess le texte sans dépendre de word_tokenize (évite l'erreur punkt_tab)"""
    if pd.isna(text) or not isinstance(text, str):
        return ""

    # Convertir en minuscules
    text = text.lower()

    # Supprimer les caractères spéciaux
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenisation simple (split sur les espaces)
    tokens = text.split()

    # Supprimer les stop words
    stop_words = set(stopwords.words('french'))
    tokens = [token for token in tokens if token not in stop_words]

    # Stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]

    return ' '.join(tokens)
```

```python
# Appliquer la transformation clean_text à la colonne 'Libellé produit'
df_cleaned['Libellé produit'] = df_cleaned['Libellé produit'].apply(clean_text)
```

```python
from sklearn.preprocessing import LabelEncoder

# Créer un encodeur pour la colonne Nature
label_encoder = LabelEncoder()

# Encoder la colonne Nature
df_cleaned['Nature_encoded'] = label_encoder.fit_transform(df_cleaned['Nature'])

# Afficher le mapping des catégories
print("Mapping des catégories Nature:")
for i, category in enumerate(label_encoder.classes_):
    print(f"{category}: {i}")
```

#### Entrainement

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np
import nltk
```

```python
# Téléchargement des données NLTK nécessaires
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df_cleaned[df_cleaned.groupby('Nature')['Nature'].transform('count') <= 2]['Nature'].value_counts() # les catégories avec moins de 3 occurrences
```

```python
# Vectorisation du texte avec TF-IDF
french_stop_words = list(stopwords.words('french'))  # Création d'une liste de stop words français
vectorizer = TfidfVectorizer(max_features=1000, stop_words=french_stop_words)
X = vectorizer.fit_transform(df_cleaned['Libellé produit'].astype(str))
# Variable cible
y = df_cleaned['Nature_encoded']
```

```python
# Division des données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split

# 1. Séparer une instance par classe à mettre dans le train set
samples_per_class = df_cleaned.groupby('Nature_encoded').head(1)
remaining_data = df_cleaned.drop(samples_per_class.index)

# 2. Split du reste (80/20, sans stratify pour éviter la même erreur)
X_remaining = vectorizer.transform(remaining_data['Libellé produit'].astype(str))
y_remaining = remaining_data['Nature_encoded']

X_rest_train, X_test, y_rest_train, y_test = train_test_split(
    X_remaining, y_remaining, test_size=0.2, random_state=42
)

# 3. Combine avec les échantillons obligatoires dans le train
X_mandatory = vectorizer.transform(samples_per_class['Libellé produit'].astype(str))
y_mandatory = samples_per_class['Nature_encoded']

# 4. Concaténation du train set final
from scipy.sparse import vstack

X_train = vstack([X_mandatory, X_rest_train])
y_train = pd.concat([y_mandatory, y_rest_train], axis=0)

# 5. Réinitialisation des index
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)

# 6. Vérification
print("Train class counts:")
print(y_train.value_counts())
print("\nTest class counts:")
print(y_test.value_counts())
```

```python
# Création et entraînement du modèle
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluation du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle: {accuracy:.4f}")
```

```python
print("\nRapport de classification:")
# Ajout du paramètre zero_division=0 pour supprimer les warnings
labels = np.unique(np.concatenate([y_test, y_pred]))
target_names = label_encoder.inverse_transform(labels)
print(classification_report(y_test, y_pred, labels=labels, target_names=target_names, zero_division=0))
```

#### Entrainement du model XGBOOST

```python
from xgboost import XGBClassifier

# Création et entraînement du modèle XGBoost
xgb_model = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    random_state=42
)

# Entraînement du modèle
xgb_model.fit(X_train, y_train)
```

```python
# Prédictions sur l'ensemble de test
y_pred_xgb = xgb_model.predict(X_test)
```

```python
# Évaluation du modèle XGBoost
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"Précision du modèle XGBoost: {accuracy_xgb:.4f}")
print(f"Précision du modèle Random Forest: {accuracy:.4f}")
print(f"Amélioration: {(accuracy_xgb - accuracy):.4f}")
```

```python
print("\nRapport de classification XGBoost:")
labels_xgb = np.unique(np.concatenate([y_test, y_pred_xgb]))
target_names_xgb = label_encoder.inverse_transform(labels_xgb)
print(classification_report(y_test, y_pred_xgb, labels=labels_xgb, target_names=target_names_xgb, zero_division=0))
```

#### Prediction des natures
